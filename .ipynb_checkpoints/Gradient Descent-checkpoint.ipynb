```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
X = 2 * np.random.rand(100,1)
```

```python
y = 4 +3 * X+np.random.randn(100,1)
```

```python
plt.scatter(X,y)
```

```python
#initializing theta
theta=np.random.randn(2,1)
```

```python
def cost(theta,X,y):
    '''
    theta: weights, they are randomly initialized
    X: features
    y: target 
    '''
    m=len(y)
    predictions=X.dot(theta)
    cost=.5*m*(predictions-y)
    return cost
```
